```python
###### 선형 리스트


## 함수 선언부



## 전역 변수부
katok = ['다현','정연','쯔위','사나','지효']


## 메인 코드부
print(katok)

# 데이터 추가 : 모모
#1단계 : 빈칸 확보
katok.append(None)
#2단계 : 마지막 칸에 넣기
katok[5] = '모모'
print(katok)

# 데이터 삽입 : 미나를 3등으로
#1단계 : 빈칸 확보
katok.append(None)
#2단계 : 한 칸씩 뒤로 이동 (모모5등~사나3등)
katok[6] = katok[5]
katok[5] = None
katok[5] = katok[4]
katok[4] = None
katok[4] = katok[3]
katok[3] = None
#3단계 : 3등 자리에 넣기
katok[3] = '미나'

print(katok)

# 데이터 삭제 : 사나
#1단계 : 데이터 지우기
katok[4] = None
#2단계 : 한 칸씩 앞으로 (5등부터~끝까지)
katok[4] = katok[5]
katok[5] = None
katok[5] = katok[6]
katok[6] = None
#3단계 : 마지막 칸 제거
del(katok[6])

print(katok)
```

```python
###### 선형 리스트 - 실무
## 함수
def add_data(friend): # 데이터 추가함수
    #1단계 : 빈칸 확보
    katok.append(None)
    klen = len(katok)
    #2단계 : 마지막 칸에 친구 넣기
    katok[klen-1] = friend

def insert_data(position,friend): # 데이터 삽입함수
    #1단계 : 빈칸 확보
    katok.append(None)
    klen = len(katok)
    #2단계 : 한 칸씩 뒤로 이동
    for i in range(klen-1,position,-1):
        katok[i] = katok[i-1]
        katok[i-1] = None
    #3단계 : 새친구 넣기
    katok[position] = friend

def delete_data(position):
    #1단계 : 데이터 지우기함수
    katok[position] = None
    klen = len(katok)
    #2단계 : 한 칸씩 앞으로 / 지울위치 다음부터 끝까지
    for i in range(position+1,klen,1):
        katok[i-1] = katok[i]
        katok[i] = None
    #3단계 : 마지막 칸 제거
    del(katok[klen-1])

## 전역
katok = []


## 메인
add_data('다현')
add_data('정연')
add_data('쯔위')
add_data('사나')
add_data('지효')
print(katok)

insert_data(3,'미나')
print(katok)

delete_data(4)
print(katok)
```

```python
###### 단순 연결 리스트 - 실무
## 함수
class Node():
    def __init__(self):
        self.data = None
        self.link = None

def printNodes(start):
    current = start
    print(current.data,end=' ')
    while(current.link != None):
        current = current.link
        print(current.data,end=' ')

def insertNode(findData,insertData):
    global memory,head,current,pre

    #Case1 : 머리 앞에 삽입
    if (findData==head.data):
        node = Node()
        node.data = insertData
        node.link = head
        head = node
        memory.append(node)
        return
    #Case2 : 중간 노드 앞에 삽입
    current = head
    while(current.link != None):
        pre = current
        current = current.link
        if (current.data==findData):
            node = Node()
            node.data = insertData
            node.link = current
            pre.link = node
            memory.append(node)
            return
    #Case3 : 없는 노드 앞에 삽입
    node = Node()
    node.data = insertData
    current.link = node
    return

def deleteNode(deleteData):
    global memory,head,current,pre

    #Case1 : 머리를 삭제
    if (deleteData==head.data):
        current = head
        head = head.link
        del(current)
        return
    #Case2 : 중간 노드를 삭제
    current = head
    while(current.link != None):
        pre = current
        current = current.link
        if (current.data==deleteData):
            pre.link = current.link
            del(current)
            return
    #Case3 : 없는 노드를 삭제
    return

def findNode(findData):
    global memory,head,current,pre
    current = head
    if (current.data==findData):
        return current
    while(current.link != None):
        current = current.link
        if (current.data==findData):
            return current
    return Node()


## 변수
memory = []
head,current,pre = None,None,None
dataArray = ['irene','seulgi','wendy','joy','yeri']


## 메인
node = Node()
node.data = dataArray[0]
head = node
memory.append(node)

for data in dataArray[1:]:
    pre = node
    node = Node()
    node.data = data
    pre.link = node
    memory.append(node)



#insertNode('irene','chaewon') # 찾을데이터, 삽입할데이터
#printNodes(head)

#insertNode('joy','kazuha')
#printNodes(head)

#insertNode('yezi','gamst')
#printNodes(head)

#deleteNode('irene') # 지울데이터
#printNodes(head)

#deleteNode('wendy')
#printNodes(head)

#deleteNode('dahyun')
#printNodes(head)

retNode = findNode('cr7')
print(retNode.data,'의 뮤비가 나옵니다. 쿵짝 쿵짝 쿵짜작 쿵짝')
```
