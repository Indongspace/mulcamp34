## GroupBy 연산
- SQL : Group By와 Logic은 동일함
- Groupby 기준컬럼 : 패턴이 있는 문자열
- Groupby 대상컬럼 : 수치 데이터
- 필요한 함수 : 집계합수(합,평균,표준편차,최대,최소,etc)

# ex 우리나라 국민의 소득 주순 비교
- 행정동 수천개 ==> 비교하는 것이 의미?
- 시군구 수백개 ==> 비교하는 것이 의미?
- 광역단위 ==> 17개로 줄임 (비교하는 것이 의미가 있음)
- 권역단위 ==> 수도권,충청권,호남권,영남권,관광특구
// 그룹바이 연산을 하겠다고 하면 데이터 가공을 해서 합쳐야됨!!
df.groupby(기준컬럼)[대상컬럼].집계

```python
df.groupby('Location')['매출액'].agg(['sum','mean','std'])
```

```python
df.groupby('Location')['매출액'].agg(['sum','mean','std']).reset_index()
```
